```python
from PIL import Image
import sys
import os
import numpy as np
from tensorflow import keras
from PIL import Image
from pathlib import Path

def crop_image(in_path, out_path, box):
    try:
        if not os.path.isfile(in_path):
            raise FileNotFoundError("Datei nicht gefunden")
        with Image.open(in_path) as img:
            cropped_img = img.crop(box)
            cropped_img.save(out_path)
            print("Bild gespeichert")
    except Exception as e:
        print(f"Fehler: {e}")

def image_size(in_path):
    if not os.path.isfile(in_path):
        raise FileNotFoundError("Datei nicht gefunden")

    try:
        with Image.open(in_path) as img:
            width, height = img.size
            return width, height
    
    except Exception as e:
        raise ValueError(f"Fehler: {e}")

width, height = image_size("test.png")
print(f"{width}, {height}")
step = 32
counter = 0
for y in range(0, height, step):
    for x in range(0, width, step):
        box = (x, y, x + step, y + step)
        out_path = f"cropped_img/title_{counter}_x{x}_y{y}.png"
        crop_image("test.png", out_path, box)
        counter +=1


# crop_image("test.png", "test_ausgabe.png", (100, 50, 400, 300))


print("Modell")

# Modell laden
filepath = Path("../models/task1a/car_cnn.keras")
model = keras.models.load_model(filepath)

def model_ex(img_path,):


    # Bild laden und vorbereiten
    img = Image.open(img_path).convert('RGB').resize((32, 32))
    img_array = np.array(img).astype('float32') / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Vorhersage
    prob = model.predict(img_array, verbose=0)[0][0]
    is_car = prob >= 0.5

    # Ausgabe
    if is_car:
        print(f"AUTO ({prob*100:.2f}%)")
    '''else:
        print(f"Was anderes({(1-prob)*100:.2f}%)")'''


width, height = image_size("test.png")
print(f"{width}, {height}")
step = 32
counter = 0
for y in range(0, height, step):
    for x in range(0, width, step):
    
        filepath_img = f"cropped_img/title_{counter}_x{x}_y{y}.png"
        model_ex(filepath_img)
        counter +=1



```
